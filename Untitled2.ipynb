```python
import zipfile
import os

# Carpeta donde se descomprimirá el dataset
extract_folder = "/content/Final_Project/Dataset/brain_tumor_dataset/"

# Crear carpeta si no existe
os.makedirs(extract_folder, exist_ok=True)

# Extraer el archivo ZIP
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

print("Dataset descomprimido en:", extract_folder)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
data_dir = "/content/Final_Project/Dataset/brain_tumor_dataset/"

img_size = (150, 150)
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=10,
    zoom_range=0.1,
    horizontal_flip=True
)

train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',   # MULTICLASE
    subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',   # MULTICLASE
    subset='validation'
)
```

```python
num_classes = 3   # porque tu dataset tiene 3 clases

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation='softmax')  # <- MULTICLASE
])

model.compile(
    loss='categorical_crossentropy',   # <- MULTICLASE
    optimizer='adam',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    train_generator,
    epochs=15,
    validation_data=validation_generator
)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Training vs Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Training", "Validation"])
plt.show()
```

```python
import random

labels = list(train_generator.class_indices.keys())

for i in range(10):
    img, label = next(validation_generator)

    pred = model.predict(img)
    pred_class = np.argmax(pred[0])
    real_class = np.argmax(label[0])

    plt.imshow(img[0])
    plt.axis("off")

    print(f"Real: {labels[real_class]} — Predicho: {labels[pred_class]}")
    plt.show()
```
